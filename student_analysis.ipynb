```python
plt.scatter(df.Attendance , df.Total_Marks)
plt.xlabel("Attendance")
plt.ylabel("Total Marks")
plt.title("Attendance vs Marks")
plt.show()
```

```python
def grade(marks):
    if marks >= 270:
        return 'A'
    elif marks >= 240:
        return 'B'
    elif marks >= 210:
        return 'C'
    else:
        return 'D'

df['Grade'] = df['Total_Marks'].apply(grade)
```

```python
df['Rank'] = df['Total_Marks'].rank(ascending=False)
```


# Insights


Most students scored highest marks in English compared to Math and Science.

Students with higher attendance generally have higher total marks, so attendance plays an important role.

A few students are top performers (Grade A), while most students fall in average performance range (Grade B and C).

Math marks show more variation, which means some students find it difficult while others perform well.

Overall, the class performance is good, with many students scoring above average marks.
